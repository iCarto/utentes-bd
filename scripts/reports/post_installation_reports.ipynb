```python
import psycopg2
conn_str = "postgresql://postgres:postgres@localhost:9001/test_arasul"
conn = psycopg2.connect(conn_str)

title = "# Explotaciones cuyo estado no coincide con el estado de la licencia"
description = "Ver ticket #2423"

display(Markdown(f"{title}\n{description}"))

sql = """
    SELECT a.gid, a.exp_name, a.exp_id, a.estado_lic, b.tipo_agua, b.estado, a.created_at, a.req_obs
    FROM utentes.exploracaos a JOIN utentes.licencias b ON a.gid = b.exploracao
    WHERE a.estado_lic != b.estado
    ORDER BY a.gid, b.tipo_agua;
"""

results = pandas.read_sql_query(sql, con=conn)
display(results)
```
